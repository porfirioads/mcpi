```python
# Importa librer√≠as necesarias
import matplotlib.pyplot as plt
from sklearn import datasets
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
# Carga el dataset
iris = datasets.load_iris()

# Convierte el dataset a pandas
df = pd.DataFrame(data=np.c_[iris['data']],
                  columns= iris['feature_names'])

# Obtiene las clases asignadas y las cambia por su etiqueta en texto
targets = pd.Series(iris.target)
labels = targets.replace(dict(enumerate(iris.target_names)))
df['label'] = labels

# Imprime las primeras filas del dataset
df.head(n=5)
```

```python
# Construye gr√°ficas de las variables
sns.pairplot(df, hue='label')
```
